```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import jpholiday
from tqdm import tqdm

# J-Quants APIのIDトークンを設定
idToken = id_token  # 必ず有効なIDトークンを設定してください
headers = {'Authorization': f'Bearer {idToken}'}

# 提出日をdatetime型に変換
df_J["提出日"] = pd.to_datetime(df_J["提出日"], format="%Y-%m-%d")

# 日付を次の平日に調整する関数
def adjust_to_next_weekday(date):
    """
    土日や祝日をスキップして次の平日を返す。
    """
    if isinstance(date, str):  # もし日付が文字列の場合はdatetimeに変換
        date = datetime.strptime(date, "%Y-%m-%d")
    while date.weekday() >= 5 or jpholiday.is_holiday(date):  # 土日または祝日ならループ
        date += timedelta(days=1)  # 次の日へ
    return date

# 各銘柄の日次終値データを取得する関数
def get_daily_close_prices(code, date):
    """
    特定の日付の終値を取得
    """
    url = f"https://api.jquants.com/v1/prices/daily_quotes?code={code}&from={date.strftime('%Y%m%d')}&to={date.strftime('%Y%m%d')}"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()["daily_quotes"]
        if data:  # データが存在する場合
            return data[0]["AdjustmentClose"]
        else:
            return None
    else:
        print(f"Error fetching data for code {code}: {response.status_code}")
        return None

# メイン処理
def fetch_stock_data(df_J):
    """
    株価データを取得して360日分に整理する。
    """
    df_J["調整日"] = df_J["提出日"].apply(lambda x: adjust_to_next_weekday(x + timedelta(days=500)))
    df_J["終値"] = None
    invalid_codes = []  # 無効な証券コードを記録

    for index, row in tqdm(df_J.iterrows(), total=df_J.shape[0], desc="株価取得進捗"):
        code = row["証券コード"]
        start_date = adjust_to_next_weekday(row["提出日"])
        end_date = row["調整日"]

        all_close_prices = []  # 収集した終値データ
        consecutive_failures = 0  # 連続失敗のカウンタ
        current_date = start_date

        # データ収集ループ
        while len(all_close_prices) < 360 and current_date <= end_date:
            close_price = get_daily_close_prices(code, current_date)

            if close_price is not None:
                all_close_prices.append(close_price)
                consecutive_failures = 0  # 成功した場合はカウンタをリセット
            else:
                consecutive_failures += 1  # 失敗した場合はカウントアップ

                # 連続8回失敗したら中止
                if consecutive_failures >= 6:
                    print(f"Data fetch aborted for code {code} due to consecutive failures.")
                    invalid_codes.append(code)
                    break

            # 次の日に進む
            current_date += timedelta(days=1)

        # 無効なコードの場合はスキップ
        if code in invalid_codes:
            continue

        # 不足している場合は追加データを取得
        while len(all_close_prices) < 360:
            additional_end_date = end_date + timedelta(days=20)
            while end_date <= additional_end_date:
                close_price = get_daily_close_prices(code, end_date)
                if close_price is not None:
                    all_close_prices.append(close_price)
                    if len(all_close_prices) >= 360:  # 360日分に達したら終了
                        break

                end_date += timedelta(days=1)

        # 360日分をリストに格納
        if len(all_close_prices) >= 360:
            df_J.at[index, "終値"] = all_close_prices[:360]

    # 無効な証券コードのデータを削除
    df_J = df_J[~df_J["証券コード"].isin(invalid_codes)]
    df_J.reset_index(drop=True, inplace=True)

    return df_J

# データフレームに適用
df_J = fetch_stock_data(df_J)
print(df_J)
```

```python
import os
import pandas as pd

# Google Drive 内の保存先パスを指定
base_path = '/content/drive/My Drive/df_savedJ-Quants_year'
file_extension = '.csv'

# ファイル番号を自動的に増加させて新しいファイル名を決定
file_number = 1
save_path = f"{base_path}{file_number}{file_extension}"

# 同じ名前のファイルがすでに存在するか確認し、存在する場合はファイル番号を増やす
while os.path.exists(save_path):
    file_number += 1
    save_path = f"{base_path}{file_number}{file_extension}"

# 同名ファイルが存在した場合、前回のデータを読み込んで結合
if file_number > 1:  # ファイル番号が2以上であれば、前回のファイルを読み込んで結合
    previous_df = pd.read_csv(f"{base_path}{file_number - 1}{file_extension}")
    df_J = pd.concat([previous_df, df_J], ignore_index=True)
    # 'docID'で重複を削除（最新のデータを優先）
    df_J.drop_duplicates(subset=['証券コード'], inplace=True)

# CSV ファイルとして保存
df_J.to_csv(save_path, index=False, encoding='utf-8-sig')  # `utf-8` で日本語対応
print(f"DataFrame を保存しました: {save_path}")
print(df_J)
```
